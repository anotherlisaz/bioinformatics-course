## Домашнее задание 4

### Выбор белка

Рассмотрим инсулин. Он понижает уровень глюкозы в крови, повышает проницаемость клетки для моносахаридов, аминокислот и жирных кислот, ускоряет гликолиз, пентозофосфатный путь и синтез гликогена в печени.

В базе Uniprot:

P01308 · INS_HUMAN - инсулин человека

P01317 · INS_BOVIN - инсулин быка

P01321 · INS_CANLF - инсулин собаки

P67970 · INS_CHICK - инсулин курицы
