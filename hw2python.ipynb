```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. Работа с таблицей

### 1.1 Чтение таблицы

```python
sns.set(style="whitegrid")

df = pd.read_csv('/home/vzolotova/homeworks/hw_2/TRB_10_7/TRB_10_7.tsv', sep='\t')

print(df.head())
```

### 1.2 Построение графика

```python
trbv_counts = df['TRBV7-9'].value_counts()
trbj_counts = df['TRBJ1-2'].value_counts()

plt.figure(figsize=(12, 6))

trbv_counts.plot(kind='bar', color='skyblue')
plt.title('Количество уникальных TRBV7-9')
plt.xlabel('TRBV7-9')
plt.ylabel('Количество')

plt.figure(figsize=(12, 6))

trbj_counts.plot(kind='bar', color='lightgreen')
plt.title('Количество уникальных TRBJ1-2')
plt.xlabel('TRBJ1-2')
plt.ylabel('Количество')

plt.tight_layout()
plt.show()
```
